# Importing Dataset

```python
import pandas as pd 
```

```python
df = pd.read_csv(r'D:\My Drive\apple_quality.csv')
df 
```

# Dataset Analysis

```python
df.isnull().sum()
```

```python
df.info()
```

# Defining X & y for Training and Testing Dataset

```python
X = df.drop(columns=['Quality','A_id'])
X 
```

```python
y = df[['Quality']]
y 
```

# Data Preprocessing

```python
y_scaled = y['Quality'].map({'good':1,'bad':0}) 
y_scaled = pd.DataFrame(y_scaled)
y_scaled 
```

# Split Dataset into Training & Testing 

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y_scaled,test_size=0.2,random_state=0) 
```

# Multiple Linear Regression 

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr = LinearRegression() 
```

```python
model = lr.fit(X_train,y_train) 
```

```python
y_pred = lr.predict(X_test)
y_pred  
```

```python
train_acc = lr.predict(X_train)
train_acc 
```

Evaultion

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np  
```

```python
print('R2_Score: ',r2_score(y_test,y_pred))
print('MSE: ',np.sqrt(mean_squared_error(y_test,y_pred))) 
```

Real Data Prediction

```python
lr.predict(pd.DataFrame([[-3.970049,-2.512336,5.346330,-1.012009,1.844900,0.329840,-0.491590]])) 
```
