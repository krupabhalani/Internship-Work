```python
import pandas as pd 
```

```python
df = pd.read_csv(r"D:\My Drive\heart.csv") 
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2) 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier 
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier() 
```

```python
clf1.fit(x_train,y_train) 
```

```python
clf2.fit(x_train,y_train) 
```

```python
y_pred1=clf1.predict(x_test)
y_pred2=clf2.predict(x_test) 
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,f1_score 
```

```python
print('Accuracy Score:')
print("LogisticRegression: ",accuracy_score(y_test,y_pred1))
print("DecisionTreeClassification: ",accuracy_score(y_test,y_pred2)) 
```

```python
print('Confusion Matrix:')
print("LogisticRegression:\n",confusion_matrix(y_test,y_pred1))
print("DecisionTreeClassification:\n",confusion_matrix(y_test,y_pred2)) 
```

```python
print('F1 Score:')
print("LogisticRegression: ",f1_score(y_test,y_pred1))
print("DecisionTreeClassification: ",f1_score(y_test,y_pred2)) 
```

```python
print('Recall Score:')
print("LogisticRegression: ",recall_score(y_test,y_pred1))
print("DecisionTreeClassification: ",recall_score(y_test,y_pred2)) 
```

```python
print('Recall Score:')
print("LogisticRegression: ",recall_score(y_test,y_pred1,average=None))
print("DecisionTreeClassification: ",recall_score(y_test,y_pred2,average=None)) 
```

```python
print('Precision Score:')
print("LogisticRegression: ",precision_score(y_test,y_pred1))
print("DecisionTreeClassification: ",precision_score(y_test,y_pred2)) 
```

```python
print('Precision Score:')
print("LogisticRegression: ",precision_score(y_test,y_pred1,average=None))
print("DecisionTreeClassification: ",precision_score(y_test,y_pred2,average=None)) 
```

```python
from sklearn.metrics import classification_report 
```

```python
print("For LR Model:")
print(classification_report(y_test,y_pred1)) 
print("\nFor DT Model:")
print(classification_report(y_test,y_pred2)) 
```
