```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r"insurance.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
X = df.drop(columns='charges')
y = df['charges']
```

```python
X
```

```python
X = pd.get_dummies(X,columns=['sex','region'],drop_first=True)
```

```python
X['smoker'] = X['smoker'].map({'yes':1,'no':0})
```

```python
X
```

```python
from sklearn.preprocessing import RobustScaler
sc = RobustScaler()
X[['age','bmi']] = sc.fit_transform(X[['age','bmi']])
X.head() 
```

# train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
X_test
```

# polinomial regression

```python
from sklearn.preprocessing import PolynomialFeatures 
```

```python
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test) 
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(X_train_poly,y_train) 
```

# prediction

```python
y_pred =lr.predict(X_test_poly)
y_pred 
```

```python
train_acc = lr.predict(X_train_poly)
train_acc 
```

# evaluation

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
r2_score(y_train,train_acc)
```

# for loop

```python
for var in range(1,5):
    poly = PolynomialFeatures(degree=var)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)
    lr = LinearRegression()
    model = lr.fit(X_train_poly,y_train)
    y_pred = lr.predict(X_test_poly)
    print(r2_score(y_test,y_pred))
```
