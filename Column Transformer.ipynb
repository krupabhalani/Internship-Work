```python
x = df[['age','gender','fever','cough','city']] 
x 
```

```python
y = df[['has_covid']]
y 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0) 
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])
x_test_fever = si.fit_transform(x_test[['fever']]) 
```

```python
x_train_fever 
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [['Mild','Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])
x_test_cough = oe.fit_transform(x_test[['cough']])
x_test_cough 
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop = 'first',sparse = False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])
x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])
x_train_gender_city 
```

```python
x_train_age = x_train.drop(columns=['gender','fever','cough','city']).values
x_test_age = x_test.drop(columns=['gender','fever','cough','city']).values
x_train_age 
```

```python
import numpy as np 
```

```python
x_train_transformed = np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis=1)
# also the test data
x_test_transformed = np.concatenate((x_test_age,x_test_fever,x_test_gender_city,x_test_cough),axis=1) 
```

```python
x_train_transformed 
```

```python
x_test_transformed 
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression() 
```

```python
a1 = x_train.iloc[:,0:1]
b1 = x_train.iloc[:,2:3] 
```

```python
a1
```

```python
c1 = b1.fillna(100)
c1 
```

```python
X_train = np.concatenate((a1,c1),axis=1)
X_train 
```

```python
a2 = x_test.iloc[:,0:1]
b2 = x_test.iloc[:,2:3]
```

```python
a2
```

```python
c2 = b2.fillna(100)
c2 
```

```python
X_test = np.concatenate((a2,c2),axis=1)
X_test 
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(x_train_transformed,y_train) 
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(x_test_transformed) 
```

```python
from sklearn.metrics import accuracy_score 
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled)) 
```

Column Transformer

```python
from sklearn.compose import ColumnTransformer 
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse=False,drop='first'),['gender','city'])
],remainder='passthrough') 
```

```python
transformer.fit_transform(x_train) 
```

```python
transformer.fit_transform(x_test) 
```
